```python
import math
from math import sqrt
from math import floor
from numpy import arange
from scipy.stats import gamma
from scipy.stats import cauchy
import matplotlib.pyplot as plt

a = 5.7
b = 2

```

```python

c = floor(a)
k = gamma.pdf(a-c, a, scale = 1/b) / gamma.pdf(a-c, c, scale = 1/(b-1))
xs = arange(0,10,0.01)

fig, ax = plt.subplots()
ax.set_xlim(0,10)
ax.set_ylim(0,1.4)
ax.plot(xs, gamma.pdf(xs,a,scale = 1/b), 'b-',linewidth=3, label='target p(x)')
ax.plot(xs, k*gamma.pdf(xs,c,scale = 1/(b-1)), 'r:',linewidth = 3, label='Gamma')
ax.legend(loc='upper right')
plt.show()
```

```python

c = (a-1)/b
g = sqrt((2*a-1)/(b*b))
k = (sqrt(2*a-1)*math.pi*math.exp(-a+1)*(a-1)**(a-1))/math.gamma(a)
fig, ax = plt.subplots()
ax.set_xlim(0,10)
ax.set_ylim(0,0.5)
ax.plot(xs, gamma.pdf(xs,a,scale = 1/b), 'b-',linewidth=3, label='target p(x)')
ax.plot(xs, k*cauchy.pdf(xs,loc=c,scale = g), 'g-',linewidth=3, label='Cauchy')
ax.legend(loc='upper right')
plt.show()
```
